```python
df = pd.read_csv('US_County_Level_Presidential_Results_08-16.csv')
```

```python
df.head()
```

```python
df['dempercent2008'] = df['dem_2008'] / df['total_2008']
df['dempercent2012'] = df['dem_2012'] / df['total_2012']
df['dempercent2016'] = df['dem_2016'] / df['total_2016']
```

```python
df.head()
```

```python
df['goppercent2008'] = df['gop_2008'] / df['total_2008']
df['goppercent2012'] = df['gop_2012'] / df['total_2012']
df['goppercent2016'] = df['gop_2016'] / df['total_2016']
```

```python
df.head()
```

```python
df['dif08'] = df['dempercent2008'] - df['goppercent2008']
df['dif12'] = df['dempercent2012'] - df['goppercent2012']
df['dif16'] = df['dempercent2016'] - df['goppercent2016']
```

```python
df.head()
```

```python
df['dif12-08'] = df['dif12'] - df['dif08']
df['dif16-08'] = df['dif16'] - df['dif08']
df['dif16-12'] = df['dif16'] - df['dif12']
```
